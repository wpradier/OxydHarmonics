### Multi Linear 3 classes :

```python
X = np.random.random((500, 2)) * 2.0 - 1.0
Y = np.array([[1, 0, 0] if -p[0] - p[1] - 0.5 > 0 and p[1] < 0 and p[0] - p[1] - 0.5 < 0 else
              [0, 1, 0] if -p[0] - p[1] - 0.5 < 0 and p[1] > 0 and p[0] - p[1] - 0.5 < 0 else
              [0, 0, 1] if -p[0] - p[1] - 0.5 < 0 and p[1] < 0 and p[0] - p[1] - 0.5 > 0 else
              [0, 0, 0]for p in X])

X = X[[not np.all(arr == [0, 0, 0]) for arr in Y]]
Y = Y[[not np.all(arr == [0, 0, 0]) for arr in Y]]
```

```python
plt.scatter(np.array(list(map(lambda elt : elt[1], filter(lambda c: Y[c[0]][0] == 1, enumerate(X)))))[:,0], np.array(list(map(lambda elt : elt[1], filter(lambda c: Y[c[0]][0] == 1, enumerate(X)))))[:,1], color='blue')
plt.scatter(np.array(list(map(lambda elt : elt[1], filter(lambda c: Y[c[0]][1] == 1, enumerate(X)))))[:,0], np.array(list(map(lambda elt : elt[1], filter(lambda c: Y[c[0]][1] == 1, enumerate(X)))))[:,1], color='red')
plt.scatter(np.array(list(map(lambda elt : elt[1], filter(lambda c: Y[c[0]][2] == 1, enumerate(X)))))[:,0], np.array(list(map(lambda elt : elt[1], filter(lambda c: Y[c[0]][2] == 1, enumerate(X)))))[:,1], color='green')
plt.show()
plt.clf()
```

```python
Y_A = np.array(list(map(lambda category: 1 if category[0] == 1 else -1, Y)))

plt.scatter(np.array(list(map(lambda elt : elt[1], filter(lambda c: Y_A[c[0]] == 1, enumerate(X)))))[:,0], np.array(list(map(lambda elt : elt[1], filter(lambda c: Y_A[c[0]] == 1, enumerate(X)))))[:,1], color='blue')
plt.scatter(np.array(list(map(lambda elt : elt[1], filter(lambda c: Y_A[c[0]] == -1, enumerate(X)))))[:,0], np.array(list(map(lambda elt : elt[1], filter(lambda c: Y_A[c[0]] == -1, enumerate(X)))))[:,1], color='red')

plt.show()
plt.clf()
```

```python
Y_B = np.array(list(map(lambda category: 1 if category[1] == 1 else -1, Y)))

plt.scatter(np.array(list(map(lambda elt: elt[1], filter(lambda c: Y_B[c[0]] == 1, enumerate(X)))))[:, 0],
            np.array(list(map(lambda elt: elt[1], filter(lambda c: Y_B[c[0]] == 1, enumerate(X)))))[:, 1], color='blue')
plt.scatter(np.array(list(map(lambda elt: elt[1], filter(lambda c: Y_B[c[0]] == -1, enumerate(X)))))[:, 0],
            np.array(list(map(lambda elt: elt[1], filter(lambda c: Y_B[c[0]] == -1, enumerate(X)))))[:, 1], color='red')

plt.show()
plt.clf()
```

```python
plt.clf()
Y_C = np.array(list(map(lambda category: 1 if category[2] == 1 else -1, Y)))

plt.scatter(np.array(list(map(lambda elt: elt[1], filter(lambda c: Y_C[c[0]] == 1, enumerate(X)))))[:, 0],
            np.array(list(map(lambda elt: elt[1], filter(lambda c: Y_C[c[0]] == 1, enumerate(X)))))[:, 1], color='blue')
plt.scatter(np.array(list(map(lambda elt: elt[1], filter(lambda c: Y_C[c[0]] == -1, enumerate(X)))))[:, 0],
            np.array(list(map(lambda elt: elt[1], filter(lambda c: Y_C[c[0]] == -1, enumerate(X)))))[:, 1], color='red')

plt.show()
plt.clf()
```

```python
linear_model_A = create_linear_model(3)
linear_model_B = create_linear_model(3)
linear_model_C = create_linear_model(3)
```

```python
train_linear_model(
      linear_model_A,
      X,
      Y_A,
      0.001,
      50000,
      True
)

train_linear_model(
      linear_model_B,
      X,
      Y_B,
      0.001,
      50000,
      True
)

train_linear_model(
      linear_model_C,
      X,
      Y_C,
      0.001,
      50000,
      True
)
```

```python
def predict_3_classes(model_A, model_B, model_C, sample):
      return np.array([
            predict_linear_model(model_A, sample, True),
            predict_linear_model(model_B, sample, True),
            predict_linear_model(model_C, sample, True)
      ])
```

```python
predict_3_classes(
      linear_model_A,
      linear_model_B,
      linear_model_C,
      np.array([-0.5, -0.5])
)
```

```python
predict_3_classes(
      linear_model_A,
      linear_model_B,
      linear_model_C,
      np.array([0., 0.75])
)
```

```python
predict_3_classes(
      linear_model_A,
      linear_model_B,
      linear_model_C,
      np.array([0.5, -0.5])
)
```

```python
destroy_linear_model(linear_model_A)
destroy_linear_model(linear_model_B)
destroy_linear_model(linear_model_C)
```
